```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel('/content/Imersão_Python_Tabela_de_acoes.xlsx', sheet_name='Principal')
df_principal
```

```python
df_total_acoes = pd.read_excel('/content/Imersão_Python_Tabela_de_acoes.xlsx', sheet_name='Total_de_acoes')
df_total_acoes
```

```python
df_ticker = pd.read_excel('/content/Imersão_Python_Tabela_de_acoes.xlsx', sheet_name='Ticker')
df_ticker
```

```python
df_empresas = pd.read_excel('/content/Imersão_Python_Tabela_de_acoes.xlsx', sheet_name='Empresas')
df_empresas
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)' : 'valor_final', 'Var. Dia (%)' : 'var_dia_percentual'}).copy()
df_principal
```

```python
df_principal['var_percentual'] = df_principal['var_dia_percentual'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_percentual'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['var_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica' : 'qnt_teorica'}).copy()
df_principal
```

```python
df_principal['Resultado'] = df_principal['var_rs'].apply(lambda x: 'Subiu' if x>0 else ('Desceu' if x<0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_empresas, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal['Faixa Etária'] = df_principal['Idade (anos)'].apply(lambda x : 'Maior que 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
# maior variação
maior = df_principal['var_rs'].max()
# menor variação
menor = df_principal['var_rs'].min()
# média variação
media = df_principal['var_rs'].mean()


media_subiu = df_principal[df_principal['Resultado']=='Subiu']['var_rs'].mean()
media_desceu = df_principal[df_principal['Resultado']=='Desceu']['var_rs'].mean()

print(f'Maior\tR$ {maior:,.2f}')
print(f'Menor\tR$ {menor:,.2f}')
print(f'Média\tR$ {media:,.2f}')
print(f'Média de quem subiu\tR$ {media_subiu:,.2f}')
print(f'Média de quem desceu\tR$ {media_desceu:,.2f}')
```

```python
df_principal_subiu = df_principal[df_principal['Resultado']== 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['var_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['var_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='var_rs', text='var_rs', title='Variação Reais por Resultado')
fig.show()
```
