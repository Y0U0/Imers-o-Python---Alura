```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python
dados = yf.download('JNJ', start='2020-01-01', end='2023-12-31')
dados = dados.reset_index()
dados
```

```python
# treinamento atÃ© o final do primeiro semestre de 2023
# teste do segundo sempestre de 2023

dados_treino = dados[dados['Date'] < '2023-07-31']
dados_teste = dados[dados['Date'] >= '2023-07-31']

```

```python
#prophet - FBPrphet
dados_prophet_treino = dados_treino[['Date', 'Close']].rename(columns={'Date':'ds', 'Close':'y'})
```

```python
# criaÃ§Ã£o e treinamento do modelo
modelo = Prophet(weekly_seasonality=True,
year_seasonality=True,
daily_seasonality=False)

modelo.add_country_holidays(country_name='US')

modelo.fit(dados_prophet_treino)
```

```python
#criaÃ§Ã£o de datas futuras, previsÃµes atÃ© o final de 2023
futuro = modelo.make_dataframe(period=100)
previsao = modelo.predict(futuro)
```

```python
plt.figure(figsize(10,8))
plt.plot(dados_treino['Date'], dados_treino['Close'], label='Dados de Treino', color='blue')
plt.plot(dados_teste['Date'], dados_teste['Close'], label='Dados de Treino', color='green')
plt.plot(previsao['ds'], previsao['yhat'], label='PrevisÃ£o', color='brown' , linestyle='--')# yhat???

plt.axvline(dados_treino['Date'].max(), color='red', linestyle='--' ,label='Inicio da PrevisÃ£o')
plt.xlabel('Data')
plt.ylabel('PreÃ§o do Fechamento')
plt.title('PrevisÃ£o de PreÃ§o de Fechemento vs Dados Reais')
plt.legend()
plt.show()
```
